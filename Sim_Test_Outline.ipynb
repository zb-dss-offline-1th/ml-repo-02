## 0-1. Def

```python
test_type = test[test['product'].str.contains('serum')] # 개인에게 할당된 type으로 변수명 설정
```

```python
def get_vector(df, idx): # test_type(df), idx
    return pd.Series(df.iloc[idx].to_numpy())[0], pd.Series((df.iloc[idx].to_numpy())[1:]) # Product Nxme, Vec

def cos_sim(x,y):
    return round(dot(x, y)/(norm(x)*norm(y)),3)

def euc_dist(x,y):   
    return round(np.sqrt(np.sum((x-y)**2)),3)

def pearson_similarity(x, y):
    return round(np.dot((x - np.mean(x)), (y - np.mean(y))) / ((np.linalg.norm(x - np.mean(x))) * (np.linalg.norm(y - np.mean(y)))),3)
```

```python
# get_vector 확인 절차
for i in range(5):
    pname, vec = get_vector(test_type, i)
    print(pname)
    print(vec[:10])
    print()
```

```python
# 유사도 함수들 확인 절차
for i in range(2):
    for j in range(i+1,2):
        pname_i, vec_i = get_vector(test_type, i)
        pname_j, vec_j = get_vector(test_type, j)

        cos = cos_sim(vec_i, vec_j)
        euc = euc_dist(vec_i, vec_j)
        pear = pearson_similarity(vec_i, vec_j)

        print(pname_i)
        print(pname_j)
        print()
        print('cos:',cos)
        print('euc:',euc)
        print('pear:',pear)
```
