## Simliarlity

```python
def get_Vector(matrix, idx):
    return matrix.iloc[idx,0], list(matrix.iloc[idx,1:])

for i in range(5):
    product_name, vec = get_Vector(matrix, i)
    print(product_name)
    print(vec[:10])
    print()
```

### Cosine Simliarlity

```python
import numpy as np
from numpy import dot
from numpy.linalg import norm

def cos_sim(A, B):
  return dot(A, B)/(norm(A)*norm(B))
```

```python
dot_lst = [(get_Vector(matrix, 100+i)[0],get_Vector(matrix, 100+i)[1]) for i in range(5)]
for i in range(len(dot_lst)):
    for j in range(i+1, len(dot_lst)):
        prod_i = dot_lst[i][0]
        prod_j = dot_lst[j][0]

        vec_i = np.array(dot_lst[i][1])
        vec_j = np.array(dot_lst[j][1])

        print(i, prod_i)
        print(j, prod_j)
        print('cos_sim:', cos_sim(vec_i, vec_j))
        print()
```

### Euclidean distance

유클리드 거리의 값이 가장 작다는 것은 유사도가 가장 높다는 의미

```python
import numpy as np

def Euc_dist(x,y):   
    return np.sqrt(np.sum((x-y)**2))
```

```python
dot_lst = [(get_Vector(matrix, 100+i)[0],get_Vector(matrix, 100+i)[1]) for i in range(5)]
for i in range(len(dot_lst)):
    for j in range(i+1, len(dot_lst)):
        prod_i = dot_lst[i][0]
        prod_j = dot_lst[j][0]

        vec_i = np.array(dot_lst[i][1])
        vec_j = np.array(dot_lst[j][1])

        print(i, prod_i)
        print(j, prod_j)
        print('Euc_dist:', Euc_dist(vec_i, vec_j))
        print()
```
