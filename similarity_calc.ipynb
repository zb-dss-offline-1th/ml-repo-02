# Similarity Calculations
- Cosine Similarity

## Cosine Similarity

```python
def filter_product(df,keyword,s,e):
    return df[df['product'].str.contains(keyword)][s:e]

def get_vector(df,idx):
    return pd.Series(df.iloc[idx].to_numpy())    

def calculate_cos_sim(x,y):
    return round(dot(x, y)/(norm(x)*norm(y)),3)
```

```python
get_vector(test,)
```

## Test Run - Essence

```python
new_df = filter_product(test,'essence',0,20)
cossim_lst = []
for i in tqdm.notebook.tqdm(range(len(new_df))):
    for j in range(len(new_df)):
        if not i==j:
            
            prod1 = get_vector(new_df,i)
            prod1vec = prod1[1:]
            prod1name = prod1[0]
            prod2 = get_vector(new_df,j)
            prod2vec = prod2[1:]
            prod2name = prod2[0]
            tmp = calculate_cos_sim(prod1vec,prod2vec)
            cossim_lst.append(tmp)
            print(prod1name,prod2name,sep="\n")
            print('CosSim:',tmp)
            print()
```

```python
print('TEST RUN AVG:',round(sum(cossim_lst)/len(cossim_lst),4))
print('TEST RUN MAX:',max(cossim_lst))
```

### ESSENCE COS SIM SUMMARY

```python
import tqdm.notebook
new_df = filter_product(test,'essence',0,30)
cos_sim_lst = []
max_cos_sim = []

for i in tqdm.notebook.tqdm(range(len(new_df))):
    for j in range(len(new_df)):
        if not i==j:
            prod1 = get_vector(new_df,i)
            prod1vec = prod1[1:]
            prod1name = prod1[0]
            prod2 = get_vector(new_df,j)
            prod2vec = prod2[1:]
            prod2name = prod2[0]
            tmp_cos_sim = calculate_cos_sim(prod1vec,prod2vec)
            cos_sim_lst.append(tmp_cos_sim)
            if tmp_cos_sim == max(cos_sim_lst):
                max_cos_sim = [prod1name,prod2name]

print('Count:',len(new_df))
print('Avg Cos Sim:',round(sum(cos_sim_lst)/len(cos_sim_lst),4))
print('Max Cos Sim:',max(cos_sim_lst))
print('Prods w/ highest Cos Sim:',max_cos_sim)
```
